**STACK DATA STRUCTURE**

*   LIFO (Last In First Out) Property

METHODS AVAILABLE
*   PUSH function: Inserts an element at the top of the stack

*   POP function: Removes an item from the top of the stack
*   isEmpty : Checks whether the stack is empty

*   isFull: Checks whether the stack is full
*   peek: get the value at the top of the stack without replacing it

IMPLEMENTATION ANALYSIS

*   Push function: O(1) (Doesn't depend on the size of the stack)
*   Pop function: O(1) (Doesn't depend on the lenght of the array since it removes just the last element)
*   isEmpty / isFull: O(1) (The len() function does not iterate over the elements of the list to calculate. It retrieves the already precalculated value from the \_\_len__ property of the object.)
*   Peek: O(1) Retrieves a single element despite the size of the stack











```python
#Implementation
class Stack:
  def __init__(self,size):
    self.stack = []
    self.size = size
    self.top = -1

  def isEmpty(self):
    if len(self.stack) == 0:
      return True
    return False

  def isFull(self):
    if len(self.stack) == self.size:
      return True
    return False

  def push(self, item):
    if self.isFull():
      print("Stack Full")
      return
    else:
      self.top += 1
      self.stack.append(item)
      return self.stack

  def pop(self):
    if self.isEmpty():
      print("Stack Empty")
      return
    else:
      removed_item = self.stack.pop(self.top)
      self.top -= 1
      return removed_item

  def print(self):
    print(self.stack)
    return

#Test cases
s =  Stack(5)

s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.push(6)
s.push(7)
s.print()
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()
s.print()
```

**CIRCULAR QUEUE**

Same mechanics as a normal queue with some improvements.
A queue would need  to be reset anytime the front pointer is not at the beginning of the array(imagine this as the space reserved for the queue in memory). A circular queue avoids this by allowing the rear pointer to go to the start of the array.

```python
#IMPLEMENTATION

class CircularQueue:
  def __init__(self, size):
    self.queue = [None] * size
    self.size = size
    self.head = self.tail = -1

  def isEmpty(self):
    if (self.head == self.tail == -1):
      return True
    else:
      return False

  def isFull(self):
    if(self.head == 0 and self.tail == self.size -1):
      return True
    elif(self.head == self.tail + 1):
      return True
    else:
      return False

  def enqueue(self, item):
    if(self.isFull()):
      print("Queue is already full")
      return 0

    if (self.isEmpty()):
      self.head = self.tail = 0
      self.queue[self.tail] = item
      return 1

    self.tail = (self.tail + 1) % self.size
    self.queue[self.tail] = item
    return 1

  def dequeue(self):
    if(self.isEmpty()):
      print("Queue is already empty")
      return 0

    temp = self.queue[self.head]
    self.queue[self.head] = None
    self.head = (self.head +1) % self.size
    return temp

  def printq(self):
    i = self.head
    temp =[]

    if(self.isEmpty()):
      print(temp)
      return 0


    while(self.queue[i] != None and len(temp) != self.size):
      temp.append(self.queue[i])
      i = (i + 1) % self.size
    print(temp)
    return 1



obj = CircularQueue(5)
obj.enqueue(1)
obj.enqueue(2)
obj.enqueue(3)
obj.enqueue(4)
obj.enqueue(5)
obj.enqueue(5)
print("Initial queue")
obj.printq()

obj.dequeue()
obj.dequeue()
print("After removing an element from the queue")
obj.printq()
```

**SIMPLE LINKED LIST**

DESCRIPTION


*   Each node in the linked list is connected in one dimension
*   Each node contains it's value, as well as a pointer to the next node.
*   Insertion is O(1). Same with deletion. Search is O(n)


OPERATIONS
*   Traversal - Going through all the elements of the linked List
*   Search - Retrieve a specified element from the linked list
*   Insertion - Insert an item into the linked list
*   Deletion - Remove an item from the linked list
*   Sort - Change the order of the liked list based on specified conditions

```python
#IMPLEMENTATION

class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def insertAtBeginning(self, data):
    data_node = Node(data)

    if self.head == None :
      self.head = data_node
    else:
      temp = self.head
      self.head = data_node
      data_node.next = temp

  def insertAtEnd(self, data):
    data_node = Node(data)

    if self.head == None:
      self.head = data_node
      return 0

    last_node = self.head

    while(last_node.next):
      last_node = last_node.next

    last_node.next = data_node

  def insertAfter(self,data,index):

    i = 0
    new_node = Node(data)
    current_node = self.head

    for i in range(index):
      if current_node is None:
        print("Given index not in the list")
        return
      else:
        current_node = current_node.next

    temp = current_node.next
    current_node.next = new_node
    new_node.next = temp


  def deleteNode(self, position):

    if self.head is None:
      print("Linked List is already empty")
      return

    temp = self.head
    if position == 0:
      self.head = temp.next
      temp = None
      return

    for i in range(position - 1):
      temp = temp.next
      if temp is None:
        break

    if temp is None:
        return

    if temp.next is None:
        return

    next = temp.next.next

    temp.next = None

    temp.next = next

  def print(self):

    output=[]
    curr_node =  self.head

    while curr_node:
      output.append(curr_node.data)
      curr_node = curr_node.next

    print(output)



l = LinkedList()

l.insertAtEnd(1)
l.insertAtEnd(2)
l.insertAtEnd(3)
l.print()
# l.deleteNode(1)
# l.print()
l.insertAtBeginning(4)
l.print()
l.insertAfter(5,1)
l.print()
l.deleteNode(4)

l.print()
```

**DOUBLY LINKED LIST**

DESCRIPTION

*   There is a previous pointer that points to the previous node in the Linked List
*   Allows traversing of the Linked Lust in both forward and backward directions
*   Shares the same methods and foundational implementation with the Singly Linked List.

```python
#IMPLEMENTATION

class Node:
  def __init__(self,data):
    self.data = data
    self.next = self.prev = None


class DoublyLinkedList:
  def __init__(self):
    self.head = None


  def insertAtBeginning(self, data):
    data_node = Node(data)

    if self.head == None :
      self.head = data_node
      return


    temp = self.head
    self.head = None
    self.head = data_node
    self.head.next = temp
    temp.prev = self.head


  def insertAtEnd(self, data):
    data_node = Node(data)

    if self.head == None:
      self.head = data_node
      return 0

    last_node = self.head

    while(last_node.next):
      last_node = last_node.next

    last_node.next = data_node
    data_node.prev = last_node


  def insertAfter(self,data,index):

    i = 0
    new_node = Node(data)
    current_node = self.head

    for i in range(index):
      if current_node is None:
        print("Given index not in the list")
        return
      else:
        current_node = current_node.next

    temp = current_node.next
    current_node.next = new_node
    new_node.prev = current_node
    new_node.next = temp
    temp.prev = new_node


  def deleteNode(self, position):

    if self.head is None:
      print("Linked List is already empty")
      return

    temp = self.head
    if position == 0:
      self.head = temp.next
      self.head.prev = None
      temp = None
      return

    for i in range(position - 1):
      temp = temp.next
      if temp is None:
        break

    if temp is None:
        return

    if temp.next is None:
        return

    next = temp.next.next

    temp.next = None

    temp.next = next
    next.prev = temp

  def print(self):

    output=[]
    curr_node =  self.head

    while curr_node:
      output.append(curr_node.data)
      curr_node = curr_node.next

    print(output)
```

**HASH TABLE**

*   This Data structure stores data in Key, value pairs.
*   When inserted into the hash table, the key is hashed by a hasing function, and then used to store the value in the memory.
*   At first, it does not make sense to hash the key, since the key itself is already unique. But hashing is important because hashing converts the key to an index in the array that is actually used to model this hash table. Hence, the need.


There are various methods to resolve hash collisions. They are as follows


1.   Chaining: If multiple keys hash to the same address, the multiple items are stored in a doubly linked list or a regular list at that same address.
2.   Open Addressing: Here, multiple items are not stored in the same index. Instead, you would have to try another address for availability. There are ways in which you probe for a free address.

  *   Linear probing: If a key hashed to an address that has been occupied, you increment to address index by 1 and try again until you have found a free address.
  *   Quadratic probing: Similar to the Linear probing. But instead of increasing the index by 1, you increase it by value generated by a quadratic function of the form c1 + c2 i ^ 2
  *   Double hashing:If a collision applies after applying a hash function, a second hash function is calculated to get another address.

```python
#HASH TABLE WITH CHAINING

class HashTableChaining:
    def __init__(self, size=100):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        for kvp in self.table[index]:
            if kvp[0] == key:
                kvp[1] = value
                return
        self.table[index].append([key, value])

    def retrieve(self, key):
        index = self._hash(key)
        for kvp in self.table[index]:
            if kvp[0] == key:
                return kvp[1]
        return None

print("Hash Table - Linear chaining")
hash_table = HashTableChaining()
hash_table.insert('apple', 'fruit')
hash_table.insert('carrot', 'vegetable')
print(hash_table.retrieve('apple'))
print(hash_table.retrieve('carrot'))


#HASH TABLE WITH LINEAR PROBING
class HashTableLinearProbing:
    def __init__(self, size=100):
        self.size = size
        self.table = [None] * size

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = (key, value)
                return
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table is full")
        self.table[index] = (key, value)

    def retrieve(self, key):
        index = self._hash(key)
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                return None
        return None


print("Hash Table - Linear Probing")
hash_table = HashTableLinearProbing()
hash_table.insert('apple', 'fruit')
hash_table.insert('carrot', 'vegetable')
print(hash_table.retrieve('apple'))
print(hash_table.retrieve('carrot'))


#HASH TABLE WITH QUADRATIC PROBING
class HashTableQuadraticProbing:
    def __init__(self, size=100, c1=1, c2=3):
        self.size = size
        self.table = [None] * size
        self.c1 = c1
        self.c2 = c2

    def _hash(self, key):
        return hash(key) % self.size

    def quad_func(self, i, index):
      return (index + self.c1 * i + self.c2 * i * i) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        i = 0
        while self.table[self.quad_func(i, index)] is not None:
            if self.table[self.quad_func(i, index)][0] == key:
                self.table[self.quad_func(i, index)] = (key, value)
                return
            i += 1
        self.table[self.quad_func(i, index)] = (key, value)

    def retrieve(self, key):
        index = self._hash(key)
        i = 0
        while self.table[self.quad_func(i, index)] is not None:
            if self.table[self.quad_func(i, index)][0] == key:
                return self.table[self.quad_func(i, index)][1]
            i += 1
        return None

print("Hash Table - Quadratic Probing")
hash_table = HashTableQuadraticProbing()
hash_table.insert('apple', 'fruit')
hash_table.insert('carrot', 'vegetable')
print(hash_table.retrieve('apple'))
print(hash_table.retrieve('carrot'))

#DOUBLE HASHING
class HashTableDoubleHashing:
    def __init__(self, size=100):
        self.size = size
        self.table = [None] * size

    def _hash1(self, key):
        return hash(key) % self.size

    def _hash2(self, key):
        return 1 + (hash(key) % (self.size - 1))

    def insert(self, key, value):
        index = self._hash1(key)
        step = self._hash2(key)
        original_index = index
        i = 0
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = (key, value)
                return
            index = (original_index + i * step) % self.size
            if index == original_index:
                raise Exception("Hash table is full")
            i += 1
        self.table[index] = (key, value)

    def retrieve(self, key):
        index = self._hash1(key)
        step = self._hash2(key)
        original_index = index
        i = 0
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (original_index + i * step) % self.size
            if index == original_index:
                return None
            i += 1
        return None

# Usage
print("Hash Table - Double Hashing")
hash_table = HashTableDoubleHashing()
hash_table.insert('apple', 'fruit')
hash_table.insert('carrot', 'vegetable')
print(hash_table.retrieve('apple'))  # Outputs: fruit
print(hash_table.retrieve('carrot'))
```

**HEAP DATA STRUCTURE**

This is a complete binary tree in which the parent node is always greater than the child node (max heap property), or vice versa(min heap property).

KEY INFO
*   Left child of a node i = 2i + 1
*   Right child of a node i = 2i + 2
*   Largest/smallest element = root of max heap/min heap respectively

METHODS
*   Heapify: Create a heap from a binary trees
*   Insert: Add an element to the heap
*   Delete: Remove an element from the heap
*   Peek: find the max or min element (for max heap and min heap respectively)

```python
#HEAP IMPLEMENTATION

class Heap:
  def __init__(self):
    self.heap=[]

  def heapify(self, index):

    largest = index
    left = (2 * index) + 1
    right = (2 * index) + 2

    if left < len(self.heap) and self.heap[left] > self.heap[largest]:
      largest = left

    if right < len(self.heap) and self.heap[right] > self.heap[largest]:
      largest = right

    if largest is not index:
      self.heap[largest], self.heap[index] = self.heap[index], self.heap[largest]
      self.heapify(largest)

  def insert(self, element):

    self.heap.append(element)
    last_non_leaf_node = len(self.heap)//2 - 1

    for i in range (last_non_leaf_node, -1, -1):
      self.heapify(i)

  def delete(self, index):

    left = 2 * index + 1
    right = 2 * index + 2
    size = len(self.heap)
    last_non_leaf_node = len(self.heap)//2 - 1

    if index >= size:
      print("Element does not exist at this position")
      return

    if left >= size:
      self.heap.pop(index)
      return

    else:
      self.heap[index], self.heap[ size - 1] = self.heap[size - 1], self.heap[index]
      self.heap.pop()
      for i in range (last_non_leaf_node, -1, -1):
        self.heapify(i)

  def print(self):
    print(self.heap)


heap = Heap()

heap.insert(1)
heap.insert(3)
heap.insert(4)
heap.insert(5)
heap.insert(10)
heap.insert(7)
heap.print()
heap.delete(0)
heap.print()
```

**TREE DATA STRUCTURE**

CHARACTERISTICS
*   Non-linear data structure
*   It consists of nodes connected by edges
*   Edge: A link between any two nodes
*   Root: Topmost node in a tree
*   Depth of a node: Number of edges from the root node to that node
*   Height of the tree: Number of edges from the deepest leaf of the tree to the root of the node.
*   Height of a node: Number of edges from the node to the deepest leaf in the tree
*   Degree of a node: This is the number of children that a node has.

TREE TRAVERSAL

This means visiting every node in the tree. There are tree main ways to do so.

1.   Depth-first search(DFS): You traverse the branches of the tree, getting to the deepest node, before moving to another branch. You do so recursively until all branches are visited.

*   Inorder: You visit the root node first, then the left, and then the right node
*   Preorder: You visit the left node first, then the root and then the right node.
*   Postorder: You visit the left node first, then the right, then the root node

2.   Breadth-first search(BFS): You traverse all the nodes at the same level before going deeper in the tree. You do this until all nodes have been traversed.

```python
#IMPLEMENTATION OF TREE TRAVERSAL

#Note that this implementation is for a binary tree traversal

class Node:
    def __init__(self, item):
        self.left = None
        self.right = None
        self.val = item

def in_order(root):

  if root:
    in_order(root.left)
    print(str(root.val), end=", ")
    in_order(root.right)

def pre_order(root):
  if root:
    print(str(root.val), end=", ")
    pre_order(root.left)
    pre_order(root.right)

def post_order(root):
  if root:
    post_order(root.left)
    post_order(root.right)
    print(str(root.val), end=", ")


root = Node(4)
root.left = Node(2)
root.right = Node(5)
root.left.left = Node(1)
root.left.right = Node(3)

print("Inorder traversal ")
in_order(root)

print("\nPreorder traversal ")
pre_order(root)

print("\nPostorder traversal ")
post_order(root)



```

**BINARY TREE**

TYPES OF BINARY TREE

1.   Full Binary Tree: This is a special type of binary tree in which each internal node has either 2 or 0 children.
2.   Perfect Binary Tree: This is a type of binary tree in which all internal node has 2 children and all the leaf nodes are on the same level.
3.   Complete Binary tree: A complete binary tree is similar to a full binary tree with the following differences.
   *   Every level must be completely filled
   *   All leaf elements must lean towards the left
   *   The last element may not necessarily have a right sibling
4.   Degenerate/pathological Binary tree: This is a binary tree in which the internal nodes have only one child, either left or right.
5.   Skewed Binary tree: A pathological binary tree with one dominating direction
6.   Balanced Binary tree: Here, the difference in height of the left and right nodes in the tree must be either 0 or 1.

BINARY SEARCH TREE

This is a binary tree that is arranged in such a way that the left node must be lesser than the root node and the right node must be greater than the root node. This must apply for all nodes in the binary tree ( that is both the left and right nodes are also BSTs )

```python
#IMPLEMENTATION OF FULL BINARY TREE

class Node:
  def __init__(self, data):
    self.left = None
    self.right = None
    self.data = data

def isFullTree(root):

  if root is None:
    return True

  if root.left is None and root.right is None:
    return True

  if root.left is not None and root.right is not None:
    return (isFullTree(root.left) and isFullTree(root.right))

  return False


root = Node(1)
root.right = Node(3)
root.left = Node(2)

root.left.left = Node(4)
root.left.right = Node(5)
root.left.right.left = Node(6)
root.left.right.right = Node(7)

print(isFullTree(root))
```

```python
# PERFECT BINARY TREE

class Node:
  def __init__(self, data):
    self.left = None
    self.right = None
    self.data = data


def calculateDepth(node):
   d = 0
   while node is not None:
    d += 1
    node = node.left
   return d

def isPerfectTree(root, depth, level=0):

  if root is None:
    return True

  if root.left is None and root.right is None:
    return depth == level + 1

  if root.left is None or root.right is None:
    return False

  return (isPerfectTree(root.left, depth, level + 1) and isPerfectTree(root.right, depth, level + 1))


root = Node(1)
root.right = Node(3)
root.left = Node(2)

root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)


if (isPerfectTree(root, calculateDepth(root))):
    print("The tree is a perfect binary tree")
else:
    print("The tree is not a perfect binary tree")
```

```python
# COMPLETE BINARY TREE

class Node:
  def __init__(self, data):
    self.left = None
    self.right = None
    self.data = data

def calculateNodesNumber(root):

  if root is None:
    return 0

  return (1 + calculateNodesNumber(root.left) + calculateNodesNumber(root.right))


def isCompleteTree(root, i, n):

  if root is None:
    return True

  if i >= n:
    return False

  return (isCompleteTree(root.left, 2 * i + 1, n) and isCompleteTree(root.right, 2 * i + 2, n))



root = Node(1)
root.right = Node(3)
root.left = Node(2)

root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

nodeCount = calculateNodesNumber(root)

if isCompleteTree(root,0, nodeCount):
    print("The tree is a complete binary tree")
else:
    print("The tree is not a complete binary tree")
```

```python
#BALANCED BINARY TREE

class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

def check_height_and_balance(node):
    """Check the height and balance of the binary tree in one traversal."""
    if node is None:
        return 0, True

    left_height, left_balanced = check_height_and_balance(node.left)
    right_height, right_balanced = check_height_and_balance(node.right)

    current_height = max(left_height, right_height) + 1
    current_balanced = left_balanced and right_balanced and abs(left_height - right_height) <= 1

    return current_height, current_balanced

def is_balanced(node):
    """Check if the binary tree is balanced."""
    _, balanced = check_height_and_balance(node)
    return balanced

# Create a balanced binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)


root.left.left.left = Node(8)
root.left.left.left.left = Node(7)

print(is_balanced(root))  # Output: True for the balanced tree; False for the unbalanced tree
```

**AVL TREE**

*   This is a self-balancing tree. It is named after Georgy Adelson-Velsky and Landis
*   It uses a balance factor to determine whether the tree is balanced or not. Balance factor = Height of left subtree - Height of right subtree


OPERATIONS
(For diagrams associated with each operation and for understanding, please visit https://www.programiz.com/dsa/avl-tree)
*   Left Rotate:  Algorithm as follows
  1.   If y has a left subtree, assign it to x's right subtree.
  2.   If x has a parent, replace x's position with y in the parent's subtree
  3.   Make y the parent of x (left  subtree)

*   Right Rotate:  Algorithm as follows
  1.   If y has a right subtree, assign it to x's left subtree.
  2.   If x has a parent, replace x's position with y in the parent's subtree
  3.   Make y the parent of x (right  subtree)

*   Left-right rotation: assuming a tree where z is the parent of y that is the parent of x;
  1.   perform a left rotation on y
  2.   perform a right rotation on z

*   Right-left rotation: assuming a tree where z is the parent of y that is the parent of x;
  1.   perform a right rotation on y
  2.   perform a left rotation on z


*   Insert: Adding a node to the tree
  1.   Locate the right position to insert the element, following the rules of a BST and insert the node. If the key already exists, exit the function
  2.   Update the height of each node
  3.   Update the balance factor of the nodes.
  4.   If the nodes are unbalanced, rebalance them according the the following principle.

  BALANCE PRINCIPLE

      a.   if balanceFactor > 1, it means that the height of the left subtree is greater than the right subtree

      > a. if the key of the newNode is lesser than the left child, perform a right rotate

      > b.else do a left-right rotation

      b.   if balanceFactor < -1, it means that the height of the right subtree is greater than the left subtree

      > a. if the key of the newNode is greater than the right child, perform a left rotate

      > b.else do a right-left rotation


*   Delete: Nodes are always deleted as leaf nodes. Here is the algorithm

  1.   Locate the correct position of the node to be deleted.
  2.   Determine how to delete the node

      a.  if the node is a leaf node, directly delete the node
      b.  if the node has only one child, swap the node to be deleted and the child element. Call the delete_node function on the child.
      c.  if the node has 2 children, swap the node to be deleted with the inorder successor ( the min node on the right subtree of the element) and call the delete_node function on that element
  
  3.   Update the height and rebalance the nodes according the balance principle.


```python
# AVL tree implementation in Python


import sys

# Create a tree node
class TreeNode(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1


class AVLTree(object):

    # Function to insert a node
    def insert_node(self, root, key):

        # Find the correct location and insert the node
        if not root:
            return TreeNode(key)
        elif key < root.key:
            root.left = self.insert_node(root.left, key)
        elif key > root.key:
            root.right = self.insert_node(root.right, key)
        else:
            print("node already exists")
            return root

        root.height = 1 + max(self.getHeight(root.left),
                              self.getHeight(root.right))

        # Update the balance factor and balance the tree
        balanceFactor = self.getBalance(root)
        if balanceFactor > 1:
            if key < root.left.key:
                return self.rightRotate(root)
            else:
                root.left = self.leftRotate(root.left)
                return self.rightRotate(root)

        if balanceFactor < -1:
            if key > root.right.key:
                return self.leftRotate(root)
            else:
                root.right = self.rightRotate(root.right)
                return self.leftRotate(root)

        return root

    # Function to delete a node
    def delete_node(self, root, key):

        # Find the node to be deleted and remove it
        if not root:
            return root
        elif key < root.key:
            root.left = self.delete_node(root.left, key)
        elif key > root.key:
            root.right = self.delete_node(root.right, key)
        else:
            if root.left is None:
                temp = root.right
                root = None
                return temp
            elif root.right is None:
                temp = root.left
                root = None
                return temp
            temp = self.getMinValueNode(root.right)
            root.key = temp.key
            root.right = self.delete_node(root.right,
                                          temp.key)
        if root is None:
            return root

        # Update the balance factor of nodes
        root.height = 1 + max(self.getHeight(root.left),
                              self.getHeight(root.right))

        balanceFactor = self.getBalance(root)

        # Balance the tree
        if balanceFactor > 1:
            if self.getBalance(root.left) >= 0:
                return self.rightRotate(root)
            else:
                root.left = self.leftRotate(root.left)
                return self.rightRotate(root)
        if balanceFactor < -1:
            if self.getBalance(root.right) <= 0:
                return self.leftRotate(root)
            else:
                root.right = self.rightRotate(root.right)
                return self.leftRotate(root)
        return root

    # Function to perform left rotation
    def leftRotate(self, z):
        y = z.right
        T2 = y.left
        y.left = z
        z.right = T2
        z.height = 1 + max(self.getHeight(z.left),
                           self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        return y

    # Function to perform right rotation
    def rightRotate(self, z):
        y = z.left
        T3 = y.right
        y.right = z
        z.left = T3
        z.height = 1 + max(self.getHeight(z.left),
                           self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        return y

    # Get the height of the node
    def getHeight(self, root):
        if not root:
            return 0
        return root.height

    # Get balance factore of the node
    def getBalance(self, root):
        if not root:
            return 0
        return self.getHeight(root.left) - self.getHeight(root.right)

    def getMinValueNode(self, root):
        if root is None or root.left is None:
            return root
        return self.getMinValueNode(root.left)

    def preOrder(self, root):
        if not root:
            return
        print("{0} ".format(root.key), end="")
        self.preOrder(root.left)
        self.preOrder(root.right)

    # Print the tree
    def printHelper(self, currPtr, indent, last):
        if currPtr != None:
            sys.stdout.write(indent)
            if last:
                sys.stdout.write("R----")
                indent += "     "
            else:
                sys.stdout.write("L----")
                indent += "|    "
            print(currPtr.key)
            self.printHelper(currPtr.left, indent, False)
            self.printHelper(currPtr.right, indent, True)


myTree = AVLTree()
root = None
nums = [33, 13, 52, 9, 21, 61, 8, 11]
for num in nums:
    root = myTree.insert_node(root, num)
myTree.printHelper(root, "", True)
key = 13
root = myTree.delete_node(root, key)
print("After Deletion: ")
myTree.printHelper(root, "", True)

print("after insertion")
myTree.insert_node(root, 8)
myTree.printHelper(root, "", True)
```

**GRAPH DATA STRUCTURE**

A graph is a collection of nodes that are connected to other nodes. Each node is called a vertex and the connection between nodes are called edges.

TERMINOLOGY
*   Adjacency: Two nodes or vertexes are adjacent if an edge connects them both.
*   Path: A series of edges that connect one vertex to another is called a path.
*   Directed graph: This is a type of graph in which the edges have a direction of traversal. For example, if there are two adjacent nodes A and B, because you can move from A to B doesn't mean you can move from B to A. (One way roads in Nigeria for instance).
*   Undirected graph: This is a type of graph in which the edges do not point in any direction.
*   Connected graph: This is a graph in which there is always a path from a vertex to any other vertex

REPRESENTATION

*   Adjacency Matrix: It a 2D V by V matrix which uses 1s and 0s to represent the presence of edges between vertexes. 1 shows the existence of an edge between the two vertexes while 0 shows the edge doesn't exist. Space complexity = O(V ^ 2)
*   Adjacency List: This represents a graph as a list of linked lists.

OPERATIONS

*   Traversal
*   Search and Retrieval
*   Insertion
*   Path finding



SPANNING TREE

This is a subgraph of a connect undirected graph, which includes all the vertices of the original graph with the minimum possible number of edges. The total number of spanning trees with n vertices that can be created from a complete graph is = n ^ (n - 2)
*   Minimum spanning tree: This is a spanning tree where the sum of the weights of each edge is a minimum as possible.

CHARACTERISTICS
*   Acyclic: The spanning tree contains no circles.
*   Connectivity: It connects all the vertices together
*   Edges: A spanning tree of graph with vertices V has exactly V - 1 edges.

STRONGLY CONNECTED COMPONENTS

A strongly connected component is a portion of a graph where all the vertexes can be accessed by each in both directions. The concerned components can be found using **Kosaraju's algorithm** or **Tarjan's algorithm**.

2.   Transpose the graph (Reverse the direction of all edges in the graph
3.   Perform a second DFS on the transposed network in the order of the stack created in step one, popping each element as they are visited.
